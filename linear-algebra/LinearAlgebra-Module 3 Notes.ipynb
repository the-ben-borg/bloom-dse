#### Exponential Functions
- Most important exponential function: 
$f(x) = e^x$
$e$ is **Euler's Number**

```python
import numpy as np
print (np.e**2)
print (np.exp(2))
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 101)

plt.plot(x, np.e**x, label="$e^x$")
plt.plot(x, 2**x, label="$2^x$")
plt.plot(x, 1**x, label="$1^x$")
plt.plot(x, 0.5**x, label="$(1/2)^x$")
plt.plot(x, (1/np.e)**x, label="$(1/e)^x$")
plt.plot(x, np.linspace(0, 0, 101), label="$x$-axis")
plt.legend()
plt.show()
```

#### Logarithmic Functions
    - Plot a pair of exponential functions and logarithmic functions below:

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-2, 2, 101)
plt.plot(x, np.exp(x), label="$e^x$")
plt.plot(x, np.log(x), label="$\ln x$")
plt.plot(x, np.linspace(0, 0, 101), label="$x$-axis")
plt.plot(np.linspace(0, 0, 101), x, label="$x$-axis")
plt.plot(x, x, label="$y=x$")
plt.ylim((-2, 2))
plt.legend()
plt.show()
```

- Plots the logarithmic function with different bases:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
step = 100
x = np.linspace(0.02, 2, step)
plt.plot(x, np.log2(x), label="$\lg x$")
plt.plot(x, np.log(x), label="$\ln x$")
plt.plot(x, [math.log(x0, 5) for x0 in x], label="$\log_5 x$")
plt.plot(x, np.log10(x), label="$\log x$")
plt.plot(x, np.linspace(0, 0, step), label="$x$-axis")
plt.plot(x, np.linspace(0, 0, step), label="$y$-axis")
plt.legend()
plt.show()
```
