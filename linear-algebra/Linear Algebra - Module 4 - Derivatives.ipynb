Graph of polynomial $f\left(x\right) = 2x^3 - 3x^2 + 9$:

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*(x**3) - 3*(x**2) + 9

def vertical_line(x):
    plt.plot(np.linspace(x, x, 101), np.linspace(0, f(2), 101), label="$x=" + str(x) + "$")
    
x = np.linspace(-1, 2, 101)
plt.plot(x, f(x), label="$2x^3 - 3x^2 + 9$")
plt.plot(x, np.linspace(0, 0, 101), label="$x$-axis")
vertical_line(0)
vertical_line(1)
plt.legend(loc="lower right")
plt.show()
```
