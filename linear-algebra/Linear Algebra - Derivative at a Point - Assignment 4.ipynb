1. The coefficient of the linear term is the ratio of two arbitrary distinct points on a line. Explain why this is so. What would happen if two points are the same?

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-2, 2, 101)
plt.style.use('ggplot')
plt.plot(x, np.linspace(0, 0, 101), label="$x$-axis")
plt.plot(np.linspace(0, 0, 101), x, label="$y$-axis")
plt.plot(x, x, label="$y=x$")
plt.plot(0, 0, 'o', label="$(0,0)$")
plt.plot(1, 1, 'o', label="$(1,1)$")
plt.ylim((-2, 2))
plt.legend()
plt.show()
```

- **Slope** describes the relationship between $x$ and $y$, so that as one value changes, the other will change in confirmity with the ratio. In a linear relationship, one arbitrary point on the line will relate to another point on the line in conformity to the slope.

- If the two points selected are the same, slope will be equal to **{}**.
