1. The coefficient of the linear term is the ratio of two arbitrary distinct points on a line. Explain why this is so. What would happen if two points are the same?

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 101)
plt.plot(x, np.linspace(0, 0, 101), label="$x$-axis")
plt.plot(np.linspace(0, 0, 101), x, label="$y$-axis")
plt.plot(x, x, label="$y=x$")
plt.plot(0, 0, 'o', label="$(0,0)$")
plt.plot(1, 1, 'o', label="$(1,1)$")
plt.ylim((-2, 2))
plt.legend()
plt.show()
```

- **Slope** describes the relationship between $x$ and $y$, so that as one value changes, the other will change in confirmity with the ratio. In a linear relationship, one arbitrary point on the line will relate to another point on the line in conformity to the slope.

- If the two points selected are the same, slope will be equal to **{}**.

2. Explain why the following plot uses `concatenate((x, x)), concatenate((sqrt(1-x*x), -sqrt(1-x*x)))` with the label `"$x^2+y^2=1$"`?

```python
x = np.linspace(-2, 2, 1001)
plt.plot(x, np.zeros(1001), label="$x$-axis")
plt.plot(np.zeros(1001), x, label="$y$-axis")
plt.plot(0, 0, 'o', label="$(0,0)$")
plt.plot(np.concatenate((x, x)), np.concatenate((np.sqrt(1-x*x), -np.sqrt(1-x*x))), label="$x^2+y^2=1$")
plt.plot(x, np.ones(1001), label="$y=1$")
plt.ylim((-2, 2))
plt.legend()
plt.show()
```

3. Functions to plot tangent and secant lines:

```python
x = np.linspace(-1, 2, 101)


def point_slope_form(x0, y0, m):
    plot(x, np.linspace(m*(-1-x0)+y0, m*(2-x0)+y0, 101), label="$y="+str(m)+"x+"+str(m*(-x0)+y0)+"$")
    

def points_form(x0, y0, x1, y1):
    point_slope_form(x0, y0, (y1-y0)/(x1-x0))
```
