### Set Operations & Sets
- Visualized with Venn Diagrams

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib_venn import venn2

S = {31, 4, 71, 65, 1773}
T = {71, 65, 1773, 19}

v = venn2([S,T], set_labels=('S', 'T'))
v.get_label_by_id('10').set_text(S - T)
v.get_label_by_id('01').set_text(T - S)
v.get_label_by_id('11').set_text(S & T)
plt.show()
```
